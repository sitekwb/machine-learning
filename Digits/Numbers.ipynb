```python
#file names
train_img = "Dataset/train-images"
train_lbl = "Dataset/train-labels"


test_img = "Dataset/test-images"
test_lbl = "Dataset/test-labels"

#init values
train_num = 60000 #number of images
test_num = 10000 #number of images
pxl_num = 28 * 28 #number of pixels in image
```

```python
import pandas as pd
import numpy as np

#Function importing labels to python
def read_labels(filename, num):
    with open(filename, 'rb') as file:
        arr = []
        i = 0
        data = file.read()
        for b in data:
            arr.append(b)
            i+=1
            if i == num:
                break
        print('i='+str(i))
        return pd.DataFrame(pd.get_dummies(pd.Series(arr), dtype=bool))
```

```python
y_test = read_labels(test_lbl, test_num)
y_test.info()
y_test.head()
```

```python
y_train = read_labels(train_lbl, train_num)
y_train.head()
```

```python
def get_images(filename, num):
    with open(filename, 'rb') as file:
        arr = np.empty((num, pxl_num), dtype=np.float64)
        data = file.read()
        col = ['Pxl'+str(i) for i in range(1, 28*28+1)]
        i = -1
        j = 0
        for pxl in data:
            if ( j%(28*28) ) == 0:
                j=0
                i+=1
                if i == num:
                    break
            arr[i][j] = pxl/255.0
            j+=1
        print("i="+str(i))
        print("j="+str(j))
    return pd.DataFrame(data=arr[:], columns=col)
```

```python
X_test = get_images(test_img, test_num)
X_test.head()
```

```python
X_train = get_images(train_img, train_num)
X_train.info()
```

```python
print("X")
X_train.info()
print("Y")
y_train.info()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

clf = MLPClassifier(hidden_layer_sizes=(100,100), alpha=0.001, solver='lbfgs', verbose=10, random_state=1)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```
